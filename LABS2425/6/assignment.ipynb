## Given the [coins.png]("coins.png") image,  
- Apply a global thresholding algorithm of your choice to split [the image]("\coins.png") into foreground and background
- Apply Otsu's Binarization
- Apply Adaptive threedholding `ADAPTIVE_THRESH_GUASSIAN_C`, set the kernel size to 11 and assign the `C` parameter a positive value.
- Apply Canny Edge Detector with `low_thresh=100` and `high_thresh=200`. Skip the gaussian smoothing.
> Plot the intensity levels from 5500:6000 along with the threshold for each image, the output should be in the following form **Except for Adaptive thresholding**:   



## 1. Global Thresholding

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("coins.png", cv2.IMREAD_GRAYSCALE)

thresh = 127

_, global_thresh = cv2.threshold(image, thresh, 255, cv2.THRESH_BINARY)

# Plot the results
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')

axes[1].plot(image.ravel()[5500:6000], label='Intensity', color='k')
axes[1].axhline(thresh, label=f'Threshold: {thresh}', color='orange', linestyle=':')
axes[1].set_ylim(-50, 400)
axes[1].legend()

axes[2].imshow(global_thresh, cmap='gray')
axes[2].set_title('Global Thresholding')

axes[3].plot(global_thresh.ravel()[5500:6000], label='Intensity', color='k')
axes[3].axhline(thresh, label=f'Threshold: {thresh}', color='orange', linestyle=':')
axes[3].set_ylim(-50, 400)
axes[3].legend()

plt.show()
```

## 2. Otsu's Binarization

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('lena.png', cv2.IMREAD_GRAYSCALE) # Change to "coins.png"

optimal_thresh, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
print(f'Threshold: {optimal_thresh}')

# Plot the results
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')

axes[1].plot(image.ravel()[5500:6000], label='Intensity', color='k')
axes[1].axhline(optimal_thresh, label=f'Threshold:{optimal_thresh}', color='orange', linestyle=':')
axes[1].set_ylim(-50, 400)
axes[1].legend()

axes[2].imshow(otsu_thresh, cmap='gray')
axes[2].set_title('Global Thresholding (Otus\'s Binarization)')

axes[3].plot(otsu_thresh.ravel()[5500:6000], label='Intensity', color='k')
axes[3].axhline(optimal_thresh, label=f'Threshold:{optimal_thresh}', color='orange', linestyle=':')
axes[3].set_ylim(-50, 400)
axes[3].legend()

plt.show()
```

## 3. Adaptive Thresholding

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('coins.png', cv2.IMREAD_GRAYSCALE)

adaptive_thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

# Plot the results
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')

axes[1].plot(image.ravel()[5500:6000], label='Intensity', color='k')
axes[1].set_ylim(-50, 400)
axes[1].legend()

axes[2].imshow(adaptive_thresh, cmap='gray')
axes[2].set_title('Adaptive Thesholding (Gaussian)')


axes[3].plot(adaptive_thresh.ravel()[5500:6000], label='Intensity', color='k')
axes[3].set_ylim(-50, 400)
axes[3].legend()


plt.show()
```

## 4. Canny Edge Detector

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('coins.png', cv2.IMREAD_GRAYSCALE)
low = 100
high = 200
canny_edges = cv2.Canny(image,low,high)

# Plot the results
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
axes = axes.ravel()

axes[0].imshow(image, cmap='gray')
axes[0].set_title('Original')

axes[1].plot(image.ravel()[5500:6000], label='Intensity', color='k')
axes[1].axhline(100, label=f'Threshold:{100}', color='orange', linestyle=':')
axes[1].axhline(200, label=f'High Threshold:{200}', color='green', linestyle=':')
axes[1].set_ylim(-50, 400)
axes[1].legend()

axes[2].imshow(canny_edges, cmap='gray')
axes[2].set_title('Canny Edge Detector')

axes[3].plot(canny_edges.ravel()[5500:6000], label='Intensity', color='k')
axes[3].axhline(100, label=f'Low Threshold:{100}', color='orange', linestyle=':')
axes[3].axhline(200, label=f'High Threshold:{200}', color='green', linestyle=':')
axes[3].set_ylim(-50, 400)
axes[3].legend()

plt.show()
```
